# Health & Wellness Sales Analysis

A cleaned and organized notebook for exploratory data analysis and visualization.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
pd.set_option("display.max_columns", None)
```

## Load Dataset

```
df = pd.read_excel("Python group 6 dataset.xlsx")
df.head()
```

## Dataset Overview

```
print("Dataset Shape:", df.shape)
df.info()
display(df.describe())
display(df.isnull().sum())
print("Duplicate Rows:", df.duplicated().sum())
```

## Categorical Variable Summary

```
categorical_columns = df.select_dtypes(include="object").columns

for column in categorical_columns:
    print(f"\n{column}")
    display(df[column].value_counts())
```

## Outlier Detection (IQR Method)

```
numeric_columns = df.select_dtypes(include="number").columns

for column in numeric_columns:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = df[(df[column] < lower) | (df[column] > upper)]

    print(f"{column}: {len(outliers)} outliers")
```

## Sales Distribution by Product Line

```
plt.figure(figsize=(8,6))
df.groupby("Product line")["Total_Cost"].sum().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90
)
plt.ylabel("")
plt.title("Sales Distribution by Product Line")
plt.tight_layout()
plt.show()
```

## Sales Trend

```
sales_trend = (
    df.groupby("Date")["Total_Cost"]
      .sum()
)

plt.figure(figsize=(10,5))
plt.plot(sales_trend.index, sales_trend.values, linewidth=2)
plt.title("Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Total Cost Distribution

```
plt.figure(figsize=(8,5))
sns.histplot(df["Total_Cost"], bins=30, kde=True)
plt.title("Distribution of Total Cost")
plt.tight_layout()
plt.show()
```

## Payment-wise Sales

```
payment_sales = (
    df.groupby("Payment", as_index=False)["Total_Cost"]
      .sum()
)

plt.figure(figsize=(6,4))
plt.bar(payment_sales["Payment"], payment_sales["Total_Cost"])
plt.title("Payment-wise Total Sales")
plt.xlabel("Payment Method")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

## Quantity Sold by Product Line

```
quantity = (
    df.groupby("Product line", as_index=False)["Quantity"]
      .sum()
)

plt.figure(figsize=(10,4))
plt.scatter(quantity["Product line"], quantity["Quantity"])
plt.xticks(rotation=30)
plt.title("Quantity Sold by Product Line")
plt.tight_layout()
plt.show()
```

## Average Total Cost by Product Line

```
average_cost = (
    df.groupby("Product line", as_index=False)["Total_Cost"]
      .mean()
)

plt.figure(figsize=(8,5))
plt.hist(average_cost["Total_Cost"], bins=10)
plt.title("Average Total Cost Distribution")
plt.xlabel("Average Total Cost")
plt.tight_layout()
plt.show()
```
